```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load data
df = pd.read_csv('DataCoSupplyChainDataset.csv', encoding='latin1')

# 2. Standardize all names and RENAME that specific long column
df.columns = [c.lower().replace(' ', '_').replace('(', '').replace(')', '') for c in df.columns]

# Rename the long name to just 'order_date' for easier use
df = df.rename(columns={'order_date_dateorders': 'order_date'})

# 3. Now 'order_date' will work perfectly!
df['order_date'] = pd.to_datetime(df['order_date'])
df['month'] = df['order_date'].dt.month_name()
df['day'] = df['order_date'].dt.day_name()

# 4. Create the Pivot Table
# values='late_delivery_risk' is the column with 1s and 0s
heatmap_data = df.pivot_table(index='month', 
                             columns='day', 
                             values='late_delivery_risk', 
                             aggfunc='mean')

# 5. Sort the Calendar
months_order = ['January', 'February', 'March', 'April', 'May', 'June', 
                'July', 'August', 'September', 'October', 'November', 'December']
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
heatmap_data = heatmap_data.reindex(index=months_order, columns=days_order)

# 6. Plot
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, cmap='YlOrRd', fmt='.2f')

plt.title('Heatmap: When are deliveries most likely to be LATE?', fontsize=15)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load and Clean
df = pd.read_csv('DataCoSupplyChainDataset.csv', encoding='latin1')
df.columns = [c.lower().replace(' ', '_').replace('(', '').replace(')', '') for c in df.columns]

# Use the correct column name we found earlier
df['order_date'] = pd.to_datetime(df['order_date_dateorders'])
df['month'] = df['order_date'].dt.month_name()

# 2. Calculate the Average Late Risk
monthly_late = df.groupby('month')['late_delivery_risk'].mean().reset_index()

# 3. SORT THE MONTHS (Crucial for Line Plots)
months_order = ['January', 'February', 'March', 'April', 'May', 'June', 
                'July', 'August', 'September', 'October', 'November', 'December']
monthly_late['month'] = pd.Categorical(monthly_late['month'], categories=months_order, ordered=True)
monthly_late = monthly_late.sort_values('month')

# 4. Create the Line Plot
plt.figure(figsize=(12, 6))

# sns.lineplot creates the trend line
# marker='o' adds dots on each month so they are easy to see
sns.lineplot(data=monthly_late, x='month', y='late_delivery_risk', marker='o', linewidth=2.5, color='teal')

# 5. Add Titles and Labels
plt.title('Monthly Trend of Late Delivery Risk', fontsize=15, fontweight='bold')
plt.xlabel('Month of the Year', fontsize=12)
plt.ylabel('Average Risk (0.0 to 1.0)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6) # Adds a background grid
plt.xticks(rotation=45) # Tilts month names for better fitting

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

df = pd.read_csv('DataCoSupplyChainDataset.csv', encoding='latin1')

# 1. Filter for ONLY orders where profit is less than 0
loss_orders = df[df['Order Profit Per Order'] < 0].copy()

# 2. Group these "loss" orders by Category to see the damage
bleeding_stats = loss_orders.groupby(['Department Name', 'Category Name']).agg({
    'Order Profit Per Order': ['sum', 'count', 'mean']
}).reset_index()

# 3. Clean up the column names for easier reading
bleeding_stats.columns = ['Department', 'Category', 'Total_Loss', 'Number_of_Loss_Orders', 'Avg_Loss_Per_Order']

# 4. Sort by the biggest total loss
bleeding_stats = bleeding_stats.sort_values(by='Total_Loss')

print("Top Categories with 'Profit Leaks':")
print(bleeding_stats.head(10))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load and Prepare (using the code we just discussed)
df = pd.read_csv('DataCoSupplyChainDataset.csv', encoding='latin1')
loss_orders = df[df['Order Profit Per Order'] < 0].copy()
bleeding_stats = loss_orders.groupby('Category Name')['Order Profit Per Order'].sum().reset_index()
bleeding_stats.columns = ['Category', 'Total_Loss']

# 2. Sort to get the top 10 biggest losses
# We sort ascending so the biggest negative numbers come first
top_10_losses = bleeding_stats.sort_values(by='Total_Loss').head(10)

# 3. Create the Plot
plt.figure(figsize=(12, 6))

# Use a nice 'Red' color to signify loss
plt.barh(top_10_losses['Category'], top_10_losses['Total_Loss'], color='#d62728')

# 4. Add labels and Title
plt.title('Top 10 Categories with Highest Total Profit Loss', fontsize=14, fontweight='bold')
plt.xlabel('Total Loss ($)', fontsize=12)
plt.ylabel('Product Category', fontsize=12)

# 5. Add a grid on the X-axis to make values easier to read
plt.grid(axis='x', linestyle='--', alpha=0.6)

# Invert the Y-axis so the #1 loser is at the top
plt.gca().invert_yaxis()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load and clean
df = pd.read_csv('DataCoSupplyChainDataset.csv', encoding='latin1')
df.columns = [c.lower().replace(' ', '_').replace('(', '').replace(')', '') for c in df.columns]

# 2. Calculate Average Profit for each Discount Rate
# This groups all orders with 10% discount, 20% discount, etc.
discount_performance = df.groupby('order_item_discount_rate')['order_profit_per_order'].mean().reset_index()

# 3. Create the Bar Graph
plt.figure(figsize=(12, 6))

# We use a color palette that goes from Blue (Profit) to Red (Loss)
# Note: 'RdYlGn_r' is a great palette for financial data
sns.barplot(data=discount_performance, 
            x='order_item_discount_rate', 
            y='order_profit_per_order', 
            palette='coolwarm')

# 4. Add the "Danger Line" (Zero Profit)
plt.axhline(0, color='black', linestyle='-', linewidth=1.5)

# 5. Formatting
plt.title('Average Profit vs. Discount Rate', fontsize=15, fontweight='bold')
plt.xlabel('Discount Rate (e.g., 0.25 = 25%)', fontsize=12)
plt.ylabel('Average Profit Per Order ($)', fontsize=12)
plt.xticks(rotation=45) # Rotates numbers so they don't overlap

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load data
df = pd.read_csv('DataCoSupplyChainDataset.csv', encoding='latin1')

# 2. Profitability by Customer Segment
# We aggregate the data to see which group (Consumer, Corporate, etc.) is most valuable
segment_stats = df.groupby('Customer Segment').agg({
    'Sales': 'sum',
    'Order Profit Per Order': 'sum',
    'Order Item Discount Rate': 'mean'
}).reset_index()

# 3. Create a Visualization: Profit vs. Discount Impact
plt.figure(figsize=(12, 6))

# Subplot 1: Which segment makes the most money?
plt.subplot(1, 2, 1)
sns.barplot(data=segment_stats, x='Customer Segment', y='Order Profit Per Order', palette='viridis')
plt.title('Total Profit by Segment')

# Subplot 2: The Correlation (Regression Plot)
# Taking a sample of 2000 rows so the chart isn't too crowded
plt.subplot(1, 2, 2)
sns.regplot(data=df.sample(2000), x='Order Item Discount Rate', y='Order Profit Per Order', 
            scatter_kws={'alpha':0.3}, line_kws={'color':'red'})
plt.title('Impact of Discounts on Profit')

plt.tight_layout()
plt.show()

# 4. Deep Insight Printout
correlation = df['Order Item Discount Rate'].corr(df['Order Profit Per Order'])
print(f"Correlation between Discount and Profit: {correlation:.4f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load Data
df = pd.read_csv('DataCoSupplyChainDataset.csv', encoding='latin1')

# 2. Corrected Column Names (Match the CSV exactly)
predictors = [
    'Order Status', 
    'Delivery Status', 
    'Shipping Mode', 
    'Late_delivery_risk', 
    'Days for shipping (real)',
    'Days for shipment (scheduled)'  # Fixed spelling/casing
]

# 3. The Loop
for i, pre in enumerate(predictors):
    plt.figure(figsize=(12, 6)) # 12,6 is usually enough for most screens
    
    # Added a palette to make it look professional
    sns.countplot(data=df, x=pre)   
    
    plt.title(f'Distribution of {pre}', fontsize=15)
    
    # Change rotation to 45 so long words like "PENDING_PAYMENT" don't crash into each other
    plt.xticks(rotation=45) 
    
    plt.tight_layout() # Prevents the labels from being cut off at the bottom
    plt.show()
```
